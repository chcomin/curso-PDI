```python
import matplotlib.pyplot as plt
import numpy as np

# Calculate histogram of the following 1D signal:
signal = [0, 10, 12, 12, 33, 0, 34, 52, 50, 70]
num_bins = 5

bin_size = (max(signal)-min(signal)+1)/num_bins
hist = np.zeros(num_bins)
num_values = len(signal)
for value in signal:
    bin_index = value/bin_size
    bin_index = int(bin_index)
    hist[bin_index] += 1
    
# Get range of each bin for visualizing the results:
bins_values = []
for i in range(num_bins+1):
    bins_values.append(i*bin_size)
print('Size of each bin: %f'%bin_size)
print('Bin values: %s'%bins_values)
print('Histogram:')
print(hist)
```

```python
plt.bar(bins_values[:-1], hist, bin_size, edgecolor='k')
```

```python
# Calculate histogram using numpy
hist, bins = np.histogram(signal, num_bins)
print('bins: %s'%bins)
print('hist: %s'%hist)
```

```python
# Calculate histogram and plot using matplotlib
plt.hist(signal, num_bins, edgecolor='k')
```

```python
# Calculate histogram of image
img = plt.imread('cameraman.tiff')

num_bins = 50             # Number of histogram bins
bin_size = 256/num_bins   # Size of each bin

hist = np.zeros(num_bins)
num_rows, num_cols = img.shape
for row in range(num_rows):
    for col in range(num_cols):
        bin_index = img[row, col]/bin_size
        bin_index = int(bin_index)
        hist[bin_index] += 1
        
# Get range of each bin for visualizing the results:        
bins_values = np.linspace(0, 256, num_bins)     
plt.bar(bins_values, hist, bin_size, edgecolor='k')
```
